```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
%matplotlib inline
```

```python
LinearRegression_score = []
power_4_score = []
power_16_score = []
```

# Number of input data: 5

```python
x = np.random.rand(5, 1) * 20
x = x.reshape(-1,1)
y = 2.3 * x - 0.2
y = np.random.normal(scale=3, size=y.shape) + y
```

```python
model_1 = LinearRegression(normalize=True)
model_1.fit(x, y)
LinearRegression_score.append(model_1.score(x, y))
```

```python
x_line = np.arange(-1, 21, 0.1).reshape(-1, 1)
y_line = model_1.predict(x_line)
plt.plot(x_line, y_line, 'r--')
plt.plot(x, y, 'bo')
plt.title('Linear Regression')
```

```python
#Adding Polynomial to the model
polynom = PolynomialFeatures(degree=4)
x_polynom = polynom.fit_transform(x)
```

```python
#Fit the polynomial Regression model
PolyReg = LinearRegression(normalize=True)
PolyReg.fit(x_polynom, y)
power_4_score.append(PolyReg.score(x_polynom, y))
```

```python
#Visualise
x_line = np.arange(-10, 40, 0.1).reshape(-1,1)
plt.plot(x_line, PolyReg.predict(polynom.fit_transform(x_line)), 'r--')
plt.plot(x, y, 'bo')
plt.title('degree 4')
```

```python
polynom = PolynomialFeatures(degree=16)
x_polynom = polynom.fit_transform(x)
```

```python
PolyReg = LinearRegression(normalize=True)
PolyReg.fit(x_polynom, y)
power_16_score.append(PolyReg.score(x_polynom, y))
```

```python
x_line = np.arange(-10, 28, 0.1).reshape(-1,1)
plt.plot(x_line, PolyReg.predict(polynom.fit_transform(x_line)), 'r--')
plt.plot(x, y, 'bo')
plt.title('degree 16')
```
